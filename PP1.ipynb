```python
def Indentation(originalPop,year,pop,diff):
    yearStr = ""
    popStr = ""
    diffStr = ""
    doubleStr = ""
    dubBin = False
    
    if len(str(year))==1:
        yearStr = "|    " + str(year) + " | "
    elif len(str(year))==2:
        yearStr = "|   " + str(year) + " | "
    else:
        yearStr = "|  " + str(year) + " | "
        
    if len(str(pop))==4:
        popStr = "        " + str(pop) + " | "
    elif len(str(pop))==5:
        popStr = "       " + str(pop) + " | "
    elif len(str(pop))==6:
        popStr = "      " + str(pop) + " | "
    elif len(str(pop))==7:
        popStr = "     " + str(pop) + " | "
    elif len(str(pop))==8:
        popStr = "    " + str(pop) + " | "
    elif len(str(pop))==9:
        popStr = "   " + str(pop) + " | "
    elif len(str(pop))==10:
        popStr = "  " + str(pop) + " | "
    elif len(str(pop))==11:
        popStr = " " + str(pop) + " | "
    elif len(str(pop))==12:
        popStr = str(pop) + " | "
    
    if len(str(diff))==4:
        diffStr = "        " + str(diff)
    elif len(str(diff))==5:
        diffStr = "       " + str(diff)
    elif len(str(diff))==6:
        diffStr = "      " + str(diff)
    elif len(str(diff))==7:
        diffStr = "     " + str(diff)
    elif len(str(diff))==8:
        diffStr = "    " + str(diff)
    elif len(str(diff))==9:
        diffStr = "   " + str(diff)
    elif len(str(diff))==10:
        diffStr = "  " + str(diff)
    elif len(str(diff))==11:
        diffStr = " " + str(diff)
    elif len(str(diff))==12:
        diffStr = str(diff)

    lineStr = yearStr+popStr+diffStr
    return lineStr
```
